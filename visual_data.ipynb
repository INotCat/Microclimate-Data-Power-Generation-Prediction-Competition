```python
#WindSpeed(m/s)
fig, ax = plt.subplots(figsize=(10, 6))

plt.scatter(df['WindSpeed(m/s)'], df['Power(mW)'], color='blue', alpha=0.5)
plt.xlabel('WindSpeed(m/s)')
plt.ylabel('Power (mW)')
plt.title('Power (mW) vs WindSpeed(m/s)')
plt.grid(True)

plt.show()
```

```python
#Pressure(hpa)
fig, ax = plt.subplots(figsize=(10, 6))

plt.scatter(df['Pressure(hpa)'], df['Power(mW)'], color='blue', alpha=0.5)
plt.xlabel('Pressure (hpa)')
plt.ylabel('Power (mW)')
plt.title('Power (mW) vs Pressure (hpa)')
plt.grid(True)

plt.show()
```

```python
#Temperature(C)
fig, ax = plt.subplots(figsize=(10, 6))

plt.scatter(df['Temperature(C)'], df['Power(mW)'], color='blue', alpha=0.5)
plt.xlabel('Temperature(C)')
plt.ylabel('Power (mW)')
plt.title('Power (mW) vs Temperature(C)')
plt.grid(True)

plt.show()
```

```python
#Humidity(%)
fig, ax = plt.subplots(figsize=(10, 6))

plt.scatter(df['Humidity(%)'], df['Power(mW)'], color='blue', alpha=0.5)
plt.xlabel('Humidity(%)')
plt.ylabel('Power (mW)')
plt.title('Power (mW) vs Humidity(%)')
plt.grid(True)

plt.show()
```

```python
#Sunlight(Lux)

fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(df['Sunlight(Lux)'], df['Power(mW)'], color='blue', alpha=0.5)
plt.xlabel('Sunlight(Lux)')
plt.ylabel('Power (mW)')
plt.title('Power (mW) vs Sunlight(Lux)')
plt.grid(True)

plt.show()
```
